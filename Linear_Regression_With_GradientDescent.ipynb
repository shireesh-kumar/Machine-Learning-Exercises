```python
#Required libraries
#Code Written with the knowledge gained from watching Andrew NG course on Coursera and Python Engineer tutorial
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
#Dataset
X , y = datasets.make_regression(n_samples=100,n_features=1,noise=20,random_state=4)
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2,random_state = 1234)
```

```python
#regressor fit
def fit(X,y,n_iters,lr):
  n_samples,n_features = X.shape
  weight = np.zeros(n_features) #theta1
  bias = 0 #thetat0

  for _ in range(n_iters):
    y_pred = np.dot(X,weight) + bias
    dw = (1/n_samples) * (np.dot(X.T,(y_pred - y)))
    db = (1/n_samples) * (np.sum(y_pred - y))

    weight = weight - (lr*dw)
    bias = bias - (lr*db)
  return weight,bias
```

```python
weight,bias = fit(X_train,y_train,1000,0.01)
```

```python
(weight,bias)
```

```python
#for training data
y_pred = np.dot(X_train,weight) + bias 
c0 = np.ones(X_train.shape[0])
plt.scatter(X_train,y_train,c = c0)
plt.plot(X_train,y_pred)
plt.show()
```

```python
y_pred = np.dot(X_test,weight) + bias
c0 = np.ones(X_test.shape[0])
plt.scatter(X_test,y_test,c = c0)
plt.plot(X_test,y_pred)
plt.show()
```
